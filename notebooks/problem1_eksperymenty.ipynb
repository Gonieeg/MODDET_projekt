```python
wszystkie = [zerowy, przy_oknie, dalej_od_okna, na_srodku, dalej_od_sciany, przy_scianie]

odleglosci = ["brak", "przy oknie", "dalej od okna", "na środku", "dalej od ściany", "przy ścianie"]

std_koncowe = [w[2] for w in wszystkie]

plt.figure(figsize=(7, 5))
plt.plot(odleglosci, std_koncowe, marker="o")
plt.title("Odchylenie standardowe temperatury końcowej")
plt.ylabel("Odchylenie standardowe [°C]")
plt.xlabel("Ustawienie grzejnika w pokoju")
plt.grid(True)
plt.show()

plt.figure(figsize=(7, 5))

for w, label in zip(wszystkie, odleglosci):
    plt.plot(range(len(w[1])), w[1], label=label, linestyle="--")

plt.title("Średnia temperatura w czasie")
plt.ylabel("Średnia temperatura [°C]")
plt.xlabel("Czas [s]")
plt.legend()
plt.grid(True)

plt.show()

```
