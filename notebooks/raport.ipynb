## Model

Do poszukiwań odpowiedzi posłuży mi model pomieszczenia wielkości 4×4m, który na północnej ścianie ma okno (wyśrodkowane i zajmujące pół ściany###). W pomieszczeniu w symetrii z oknem umieszczam grzejnik rozmiarów 10×100cm (i wysokości 60, ale tego nie widać w symulacji) i testuję, jak temperatura w pokoju zachowuje się w zależności od jego odległości od okna. Mierzę to za pomocą średniej oraz odchylenia standardowego z temperatury panującej we wnętrzu pomieszczenia (nie licząc chłodzących ścian i okna). 

Dla zachowania poprawności eksperymentu, każdy zaczyna się w takich samych warunkach początkowych: na zewnątrz panuje temperatura $0^o$C, a wewnątrz $20^\circ$C. Grzejnik jest włączony na maksymalne pokrętło - czyli grzeje, aż temperatura osiągnie 28$20^\circ$C (zazwyczaj przez całą symulację). Symulacje są prowadzone przez tyle samo czasu - 10800 sekund, czyli 3 godziny.

Do modelowania zmian temperatury pomieszczenia służy układ równań:

$$ \begin{equation} \left\{ \begin{aligned} \frac{\partial u}{\partial t} &= \alpha \Delta u + u \cdot \frac{P \cdot r}{p \cdot A \cdot c}\,\Theta(x,u), && x \in R_i,\; t \in [0,T],\; i \in \{1,\dots,N_{\text{rooms}}\}, \\[8pt] \frac{\partial u}{\partial \vec{n}} &= -\frac{\lambda_{\text{window}}}{\lambda_{\text{air}}} \left(u - u_{\text{z}}\right), && x \in W_i,\; t \in [0,T],\; i \in \{1,\dots,N_{\text{windows}}\}, \\[8pt] \frac{\partial u}{\partial \vec{n}} &= -\frac{\lambda_{\text{wall}}}{\lambda_{\text{air}}} \left(u - u_{\text{z}}\right), && x \in W_i,\; t \in [0,T],\; i \in \{1,\dots,N_{\text{walls}}\}, \\[8pt]
u(x,0) &= 20,
&& x \in \Omega.
\end{aligned}
\right.
\end{equation} $$

gdzie $\Theta(x,u)$:

$$
\begin{equation}
\Theta(x,u) =
\begin{cases}
1, & \text{jeśli } \displaystyle
\frac{1}{|R_i|}\int_{R_i} u(x,t)\,\mathrm{d}x \le S_i, \\[10pt]
0, & \text{w przeciwnym razie}.
\end{cases}
\end{equation}
$$
